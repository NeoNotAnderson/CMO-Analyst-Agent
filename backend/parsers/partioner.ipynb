```python
elements = partition_pdf(
    filename="JPM03_summary00.pdf",
    strategy='hi_res',
    infer_table_structure=True
)
```

```python
with open('output.txt', 'w') as f:
    print(len(elements), file=f)
    print('.........', file=f)
    for element in elements:
        if element.category == 'Title':
            print(element.text, file=f)
        elif element.category == 'Text':
            parts = element.split('\n')
            for part in parts:
                print(part[ : 50], file=f)
        else:
            print(element.category, file=f)
            print(element.text[:50], file=f)
        print(f'......{element.metadata.page_number}.....', file=f)
```
